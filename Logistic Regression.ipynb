```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Values for plotting the standard normal CDF
x = np.linspace(-3, 3, 1000)
cdf = stats.norm.cdf(x)

# Plotting the CDF
plt.figure(figsize=(8, 5))
plt.plot(x, cdf, label='Standard Normal CDF', color='blue')
plt.title('CDF of the Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Φ(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm

# Generate synthetic data
np.random.seed(0)
n = 100
X = np.random.normal(size=n)
beta = 1.5
Y = (np.random.normal(size=n) < beta * X).astype(int)

# Add a constant term to the predictors
X_const = sm.add_constant(X)

# Fit the probit model
probit_model = sm.Probit(Y, X_const)
probit_result = probit_model.fit()

# Display the results
print(probit_result.summary())
```
