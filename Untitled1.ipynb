```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have retrieved the data from the database and stored it in a DataFrame called 'crime_data'
# The DataFrame 'crime_data' should contain columns: 'week_number' and 'crime_type'
crime_data = pd.read_csv("Q1.csv")

# Grouping data by week_number and counting the occurrences of each crime type
crime_type_counts = crime_data.groupby('crime_type')['crime_count'].sum()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Crime Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have retrieved the data from the database and stored it in a DataFrame called 'crime_data'
# The DataFrame 'crime_data' should contain columns: 'crime_type' and 'crime_count'

# Grouping data by crime type and summing the counts
crime_type_counts = crime_data.groupby('crime_type')['crime_count'].sum()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
crime_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have retrieved the data from the database and stored it in DataFrame df
df = pd.read_csv("Q2.csv") 
# Plotting the bar chart
plt.figure(figsize=(12, 8))

# Plotting population density
plt.bar(df['precinct_code'], df['total_population_density'], color='skyblue', label='Population Density')

# Plotting precinct area
plt.bar(df['precinct_code'], df['total_areas'], color='lightgreen', label='Precinct Area')

# Plotting officer count
plt.bar(df['precinct_code'], df['total_officers'], color='salmon', label='Officer Count')

plt.title('Comparison of Population Density, Precinct Area, and Officer Count by Precinct')
plt.xlabel('Precinct Code')
plt.ylabel('Value')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have retrieved the data from the database and stored it in DataFrame df
df = pd.read_csv("Q3.csv") 

# Plotting the bar chart
plt.figure(figsize=(12, 8))

# Grouping data by day part and crime type and summing the counts
crime_counts_by_daypart = df.groupby(['time_part', 'crime_type'])['num_crimes'].sum().reset_index()

# Plotting the bar chart
sns.barplot(data=crime_counts_by_daypart, x='time_part', y='num_crimes', hue='crime_type', palette='muted')
plt.title('Distribution of Most Prevalent Crime Types Throughout the Day')
plt.xlabel('Day Part')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Crime Type', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (assuming df is already available and properly structured)
# df = pd.DataFrame({
#     'area_name': ['Area1', 'Area2', 'Area1', 'Area2', 'Area3', 'Area3', 'Area3'],
#     'time_part': ['Morning', 'Morning', 'Afternoon', 'Afternoon', 'Night', 'Night', 'Morning'],
#     'num_crimes': [15, 20, 25, 10, 5, 8, 12]
# })

# Ensure df contains the necessary columns: 'area_name', 'time_part', 'num_crimes'
# You can preprocess df to ensure it's in the correct format if needed
df = pd.read_csv("Q4.csv") 

# Grouping data by area_name and time_part and summing the counts
area_time_crimes = df.groupby(['area_name', 'time_part'])['max_crimes'].sum().reset_index()

# Pivot the data to have time parts as columns for easier plotting
pivot_df = area_time_crimes.pivot(index='area_name', columns='time_part', values='max_crimes').fillna(0)

# Plotting the bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title('Distribution of Maximum Crimes Reported Across Day Parts by Area')
plt.xlabel('Area Name')
plt.ylabel('Number of Crimes')
plt.legend(title='Time Part')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Q8.csv") 

df_melted = pd.melt(df, id_vars=['area_name'], 
                    value_vars=['footage_available', 'footage_not_available'],
                    var_name='footage_status', value_name='num_crimes')

# Plotting the bar chart
plt.figure(figsize=(14, 8))

# Using seaborn to create a bar chart
sns.barplot(data=df_melted, x='area_name', y='num_crimes', hue='footage_status', palette='muted')

# Adding titles and labels
plt.title('Crimes Committed by Relation of Victim/Strangers')
plt.xlabel('Area Name')
plt.ylabel('Number of Crimes')
plt.legend(title='Footage Status', labels=['Footage Available', 'Footage Not Available'])
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Q7.csv") 

# Melt the dataframe to have a long format suitable for seaborn
df_melted = pd.melt(df, id_vars=['area_name'], 
                    value_vars=['cctv_count', 'num_crimes'],
                    var_name='metric', value_name='value')

# Plotting the grouped bar chart
plt.figure(figsize=(14, 8))

# Using seaborn to create a grouped bar chart
sns.barplot(data=df_melted, x='area_name', y='value', hue='metric', palette='muted')

# Adding titles and labels
plt.title('Distribution of Area and CCTV Count in Relation to Crime Rate')
plt.xlabel('Area Name')
plt.ylabel('Count')
plt.legend(title='Metric', labels=['CCTV Count', 'Number of Crimes'])
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Q8.csv") 
# Set the area_name as the index for easy plotting
df.set_index('area_name', inplace=True)

# Plotting the bar chart
plt.figure(figsize=(14, 8))

# Plotting the total count of cases
df['total_cases'] = df['footage_available'] + df['footage_not_available']
plt.bar(df.index, df['total_cases'], label='Total Cases')

# Plotting the count of cases with CCTV present
plt.bar(df.index, df['footage_available'], label='CCTV Present')

# Plotting the count of cases with no CCTV footage available
plt.bar(df.index, df['footage_not_available'], label='CCTV Not Present')

# Adding titles and labels
plt.title('Distribution of CCTV Footage Availability at Crime Scenes by Area')
plt.xlabel('Area Name')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Q9.csv") 

# Sample Data
df = pd.DataFrame({
    'crime_type': ['Robbery', 'Burglary', 'Assault', 'Fraud'],
    'offender_relation': ['Stranger', 'Acquaintance', 'Family', 'Stranger'],
    'num_crimes': [50, 40, 30, 20]
})

# Filter the data based on offender relation (e.g., 'Stranger')
stranger_crimes = df[df['offender_relation'] == 'Stranger']

# Plotting the bar chart
plt.figure(figsize=(10, 6))

# Plotting the number of cases reported for each crime type
plt.bar(stranger_crimes['crime_type'], stranger_crimes['num_crimes'])

# Adding titles and labels
plt.title('Crimes Committed by Relation of Victim/Strangers')
plt.xlabel('Crime Type')
plt.ylabel('Number of Cases Reported')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
